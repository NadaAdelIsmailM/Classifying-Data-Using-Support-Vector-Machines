#Importing important libiraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

#Loading our datasets

```python
#Aggregation.txt
Aggregation_data = np.loadtxt('Aggregation.txt',skiprows=7)#skip the header part
Aggregation_X, Aggregation_y = Aggregation_data[:, :-1], Aggregation_data[:, -1]
#Compound.txt
Compound_data = np.loadtxt('Compound.txt',skiprows=7)#skip the header part
Compound_X, Compound_y = Compound_data[:, :-1], Compound_data[:, -1]
#Flame.txt
Flame_data = np.loadtxt('Flame.txt',skiprows=7)#skip the header part
Flame_X, Flame_y = Flame_data[:, :-1], Flame_data[:, -1]
#Jain.txt
Jain_data = np.loadtxt('Jain.txt',skiprows=7)#skip the header part
Jain_X, Jain_y = Jain_data[:, :-1], Jain_data[:, -1]
#Pathbased.txt
Pathbased_data = np.loadtxt('Pathbased.txt',skiprows=7)#skip the header part
Pathbased_X, Pathbased_y = Pathbased_data[:, :-1], Pathbased_data[:, -1]
#Spiral.txt
Spiral_data = np.loadtxt('Spiral.txt',skiprows=7)#skip the header part
Spiral_X, Spiral_y = Spiral_data[:, :-1], Spiral_data[:, -1]
```

#Visualize the datasets

```python
#Aggregation dataset
plt.scatter(Aggregation_X[:, 0], Aggregation_X[:, 1], c=Aggregation_y, cmap=plt.cm.Paired)
plt.title('Aggregation Dataset')
plt.show()
#Compound dataset
plt.scatter(Compound_X[:, 0], Compound_X[:, 1], c=Compound_y, cmap=plt.cm.Paired)
plt.title('Compound Dataset')
plt.show()
#Flame dataset
plt.scatter(Flame_X[:, 0], Flame_X[:, 1], c=Flame_y, cmap=plt.cm.Paired)
plt.title('Flame Dataset')
plt.show()
#Jain dataset
plt.scatter(Jain_X[:, 0], Jain_X[:, 1], c=Jain_y, cmap=plt.cm.Paired)
plt.title('Jain Dataset')
plt.show()
#Pathbased dataset
plt.scatter(Pathbased_X[:, 0], Pathbased_X[:, 1], c=Pathbased_y, cmap=plt.cm.Paired)
plt.title('Pathbased Dataset')
plt.show()
#Spiral dataset
plt.scatter(Spiral_X[:, 0], Spiral_X[:, 1], c=Spiral_y, cmap=plt.cm.Paired)
plt.title('Spiral Dataset')
plt.show()
```

# Split the datasets into training and testing sets

```python
#Aggregation dataset
Aggregation_X_train, Aggregation_X_test, Aggregation_y_train, Aggregation_y_test = train_test_split(Aggregation_X, Aggregation_y, test_size=0.2, random_state=42)

#Compound dataset
Compound_X_train, Compound_X_test, Compound_y_train, Compound_y_test = train_test_split(Compound_X, Compound_y, test_size=0.2, random_state=42)

#Flame dataset
Flame_X_train, Flame_X_test, Flame_y_train, Flame_y_test = train_test_split(Flame_X, Flame_y, test_size=0.2, random_state=42)

#Jain dataset
Jain_X_train, Jain_X_test, Jain_y_train, Jain_y_test = train_test_split(Jain_X, Jain_y, test_size=0.2, random_state=42)

#Pathbased dataset
Pathbased_X_train, Pathbased_X_test, Pathbased_y_train, Pathbased_y_test = train_test_split(Pathbased_X, Pathbased_y, test_size=0.2, random_state=42)

#Spiral dataset
Spiral_X_train, Spiral_X_test, Spiral_y_train, Spiral_y_test = train_test_split(Spiral_X, Spiral_y, test_size=0.2, random_state=42)
```

# SVM with different kernels and regularization

```python
kernels = ['linear', 'poly', 'rbf']
regularizations = [0.00000000001, 10000]
Datasets=["Aggregation","Compound","Flame","Jain","Pathbased","Spiral"]
X_trains=[Aggregation_X_train,Compound_X_train,Flame_X_train,Jain_X_train,Pathbased_X_train,Spiral_X_train]
y_trains=[Aggregation_y_train,Compound_y_train,Flame_y_train,Jain_y_train,Pathbased_y_train,Spiral_y_train]
X_tests=[Aggregation_X_test,Compound_X_test,Flame_X_test,Jain_X_test,Pathbased_X_test,Spiral_X_test]
y_tests=[Aggregation_y_test,Compound_y_test,Flame_y_test,Jain_y_test,Pathbased_y_test,Spiral_y_test]
for i in range(len(Datasets)):
  for kernel in kernels:
      for C in regularizations:
          # Create and train the SVM model
          model = SVC(kernel=kernel, C=C, verbose=False)
          model.fit(X_trains[i], y_trains[i]);

          # Make predictions on the testing set
          y_pred = model.predict(X_tests[i])

          # Evaluate the model
          accuracy = accuracy_score(y_tests[i], y_pred)
          report = classification_report(y_tests[i], y_pred)

          plt.scatter(X_tests[i][:, 0], X_tests[i][:, 1], c=y_pred, cmap=plt.cm.Paired, marker='x')
          plt.title(f'{Datasets[i]} SVM with {kernel} kernel, {"regularization" if C==max(regularizations) else "no regularization"}, Accuracy={accuracy:.2f}')
          plt.show()

          # Print results
          print(f"SVM with {kernel} kernel, C={C}")
          print("Accuracy:", accuracy)
          print("Classification Report:\n", report)
          print("_-_-" * 50)
```
